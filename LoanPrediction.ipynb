**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action="ignore")
```

**Importing train file**

```python
train_data= pd.read_csv("train_ctrUa4K.csv")
train_data.head()
```

```python
train_data.describe(include="all")
```

```python
train_data=train_data.dropna()
train_data.describe()
```

```python
train_data=train_data.drop(["Loan_ID"],axis=1)
train_data
```

```python
def label_encode(sex):
  if sex=="Male":
    return 1
  else:
    return 0
train_data["Gender"]= train_data["Gender"].apply(label_encode)
```

```python
train_data
```

```python
def is_married(marry):
  if marry=="Yes":
    return 1
  else:
    return 0
train_data["Married"]=train_data["Married"].apply(is_married)
```

```python
train_data
```

```python
def is_graduate(grad):
  if grad=="Graduate":
    return 1
  else:
    return 0
train_data["Education"]=train_data["Education"].apply(is_graduate)
train_data
```

```python
def is_selfemp(emp):
  if emp=="Yes":
    return 1
  else:
    return 0
train_data["Self_Employed"]= train_data["Self_Employed"].apply(is_selfemp)
train_data
```

```python
def prop_area(prop):
  if prop=="Rural":
    return 0
  elif prop == "Semiurban":
    return 1
  else:
    return 2
train_data["Property_Area"]= train_data["Property_Area"].apply(prop_area)
train_data        
```

```python
def status(loan):
  if loan=="Y":
    return 1
  else:
    return 0
train_data['Loan_Status'] = train_data['Loan_Status'].apply(status)
train_data   
```

```python

def depend(dep):
  if dep=="3+":
    return 3
  else:
    return dep
train_data["Dependents"]= train_data["Dependents"].apply(depend)
train_data

```

**Segregating Independent & dependent variable**

```python
train_X= train_data.drop(["Loan_Status"], axis=1)
train_y= train_data["Loan_Status"]
```

```python
train_X.head()
```

```python
train_y.head()
```

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(train_X,train_y)
```

```python
predict_train=svc.predict(train_X)
predict_train
```

```python
from sklearn.metrics import accuracy_score
accuracy_train = accuracy_score(train_y,predict_train)
```

```python
accuracy_train
```
